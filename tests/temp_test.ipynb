```python
# the movie column in the img df is a string instead of a list of strings
# we need to convert it to a list of strings and remove ' characters from the resulting strings
# For example: "['tt4003966']" -> ['tt4003966']

img['movie'] = img['movie'].apply(lambda x: x[2:-2].split("', '"))
```
