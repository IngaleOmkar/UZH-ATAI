```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/movies.csv")
df2 = pd.read_csv("data/archive/action.csv")
df3 = pd.read_csv("data/archive/adventure.csv")
df4 = pd.read_csv("data/archive/animation.csv")
df5 = pd.read_csv("data/archive/biography.csv")
df6 = pd.read_csv("data/archive/crime.csv")
df7 = pd.read_csv("data/archive/family.csv")
df8 = pd.read_csv("data/archive/fantasy.csv")
df9 = pd.read_csv("data/archive/film-noir.csv")
df10 = pd.read_csv("data/archive/history.csv")
df11 = pd.read_csv("data/archive/mystery.csv")
df12 = pd.read_csv("data/archive/romance.csv")
df13 = pd.read_csv("data/archive/scifi.csv")
df14 = pd.read_csv("data/archive/thriller.csv")
df15 = pd.read_csv("data/archive/war.csv")
df16 = pd.read_csv("data/archive/sports.csv")
```

```python
# Merge df2 through df16 by columns and remove duplicates
df = pd.concat([df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16], ignore_index=True)
```

```python
df['genre'] = df['genre'].str.split(',')
```

```python
df['star'] = df['star'].str.split('\n')
```

```python
# drop the columns: 'movie_id', 'certificate', 'runtime', 'rating', 'description', 'director_id', 'star_id', 'votes', 'gross(in $)
df.drop(['movie_id', 'certificate', 'runtime', 'rating', 'description', 'director_id', 'star_id', 'votes'], axis=1, inplace=True)
```

```python
df.drop(['gross(in $)'], axis=1, inplace=True)
```

```python
# only keep rows with unique movie_name in df
df = df.drop_duplicates(subset='movie_name', keep='first')
```

```python
# get number of rows where movie_name is na/nan or ""
df['movie_name'].isna().sum()
```

```python
# drop rows where movie_name is na/nan or ""
df = df.dropna(subset=['movie_name'])
```

```python
df.to_csv('data/movies2.csv', index=False)
```
